### Creating Plots For Project

```
import pandas as pd
import altair as alt
!pip install wordcloud
from wordcloud import WordCloud
from wordcloud import ImageColorGenerator
from wordcloud import STOPWORDS
import matplotlib.pyplot as plt
```

```
def create_word_plots(genre):
    data_dump = pd.read_csv('pure_genre_data_w_clean_lyrics.csv')
    data_dump = data_dump.dropna()
    data_dump['lyric_clean']= data_dump['lyric_clean'].apply(str)

    alt_df = data_dump[data_dump['genre'] == genre]

    text = " ".join(i for i in alt_df['lyric_clean'])
    
    #stopwords = set(STOPWORDS)
    wordcloud = WordCloud(background_color="white").generate(text)
    plt.figure(figsize=(35,10))
    plt.title("WordCloud for Lyrics in " + genre + " Genre",fontsize=20)
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()
    
```

```
def create_attribute_plots(genre):
    data_dump = pd.read_csv('pure_genre_data_w_clean_lyrics.csv')
    data_dump = data_dump.dropna()
    avg_df = data_dump.groupby('genre')[['danceability', 'energy', 'key', 'loudness', 'mode',
    'speechiness','acousticness','instrumentalness','liveness','valence','tempo',
    'duration_ms','time_signature']].mean().reset_index()



    fixed = avg_df[avg_df['genre'] == genre]
    fixed['loudness'] =(10 ** (fixed['loudness']/10)) 
    fixed.plot(kind = 'bar', figsize=(20, 10))
    plt.title( "Average Track Attributes for " + genre,fontsize=20)
    plt.ylabel('Log-Scaled Track Attributes', fontsize=16)
    plt.yticks(fontsize=14)
    plt.legend(loc=1, prop={'size': 15})
    plt.yscale("log")
    plt.show()
```

```
create_attribute_plots('country')
```

```
create_attribute_plots('heavy-metal')
```

```
create_word_plots('country')
```

```
create_word_plots('heavy-metal')
```
