### Creating Plots For Project

```
import pandas as pd
import altair as alt
!pip install wordcloud
from wordcloud import WordCloud
from wordcloud import ImageColorGenerator
from wordcloud import STOPWORDS
import matplotlib.pyplot as plt
```

```
data_dump = pd.read_csv('pure_genre_data_w_clean_lyrics.csv')
data_dump.dropna()
data_dump['lyric_clean']= data_dump['lyric_clean'].apply(str)

alt_df = data_dump[data_dump['genre'] == 'alt-rock']

text = " ".join(i for i in alt_df['lyric_clean'])

text
```

```
alt_df_mean = alt_df.mean(axis = 0, skipna = False)
#df_mean_frame = alt_df_mean.to_frame()
#df_mean_frame['duration_min'] = df_mean_frame['duration_ms'] * 0.001
#df_mean_frame.reset_index()



fixed = alt_df_mean
plt.figure(figsize=(10,10)) 
fixed.plot(kind = 'bar', title = 'Alt-Rock Attributes', color="red")
plt.yscale("log") 
plt.show()
```

```
avg_df = data_dump.groupby('genre')[['danceability', 'energy', 'key', 'loudness', 'mode',
 'speechiness','acousticness','instrumentalness','liveness','valence','tempo',
 'duration_ms','time_signature']].mean().reset_index()

avg_df_plt = avg_df[['danceability', 'energy', 'key', 'loudness', 'mode',
 'speechiness','acousticness','instrumentalness','liveness','valence', 'genre']]

f, ax = plt.subplots(3, 2, figsize=(12, 15))
avg_df_plt.groupby('genre').plot(kind='bar')
```

```
##Word-Cloud for genres
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)
plt.figure( figsize=(35,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
